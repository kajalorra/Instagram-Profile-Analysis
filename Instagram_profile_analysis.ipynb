```python
#Category visulization
plt.figure(figsize=(10,5))

df["Category"].value_counts().plot(kind="bar")

plt.title("Instagram Categories")

plt.xlabel("Category")

plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show() 
```

```python
#Followers distribution
plt.figure(figsize=(8,5))

plt.hist(df["Followers"], bins=20)

plt.title("Followers Distribution")

plt.xlabel("Followers")

plt.ylabel("Profiles")

plt.show()
```

```python
#Following distribution
plt.hist(df["Following"], bins=20)

plt.title("Following Distribution")

plt.xlabel("Following")

plt.ylabel("Profiles")

plt.show()
```

```python
#Posts distribution
plt.hist(df["Posts"], bins=30)

plt.title("Posts Distribution")

plt.xlabel("Posts")

plt.ylabel("Profiles")

plt.show()
```

```python
#Top 10 Profiles by Followers
top10 = df.sort_values("Followers", ascending=False).head(10)

top10
```

```python
#  Visualize Top 10 Profiles
plt.figure(figsize=(12,5))

plt.bar(top10["Username"], top10["Followers"])

plt.xticks(rotation=60)

plt.title("Top 10 Instagram Profiles")

plt.show()
```

```python
# Create follower ratio
df["Follower_Ratio"] = df["Followers"] / df["Following"]
```

```python
# Highest Follower Ratio
df.sort_values("Follower_Ratio", ascending=False).head(10)
```

```python
 # Save Clean Dataset
df.to_csv("clean_instagram_profiles.csv", index=False)
```
